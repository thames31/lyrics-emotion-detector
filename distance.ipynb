```python
import pandas as pd
from sklearn.metrics import pairwise_distances
import numpy as np
from sklearn.neighbors import BallTree
```

```python
en_songs = pd.read_csv('data/en_songs3.csv')
en_songs.head()
```

```python
df = en_songs.loc[:, en_songs.columns.intersection(['track_artist','track_name','danceability','energy', 'speechiness', 'acousticness', 'duration_ms', 'tempo', 'instrumentalness', 'cluster_label'])]
df
```

```python
distances = pairwise_distances(df[['danceability', 'energy', 'speechiness', 'acousticness','tempo', 'instrumentalness', 'cluster_label']])
```

```python
distances = np.round(distances,2)
```

```python
distances.max()
```

```python
np.fill_diagonal(distances,180)
```

```python
df = pd.DataFrame(distances, columns=list(en_songs['track_name']))
```

```python
df['id'] = en_songs['track_id']
df
```

```python
df = df.set_index('id', drop=True)
```

```python
df
```

```python
song_name = 'I Feel Alive'
dl = df.sort_values(by=song_name)
```

```python
ids = list(dl.head(5).index)
links = []

for id in ids:
    link = "https://open.spotify.com/track/" + id
    links.append(link)

links
```

```python
btree = BallTree(df, metric='euclidean')
```

```python
dist, ind = btree.query(df[:1], k=6) 
```

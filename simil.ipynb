```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
en_songs = pd.read_csv('data/en_songs3.csv')
df = en_songs.loc[:, en_songs.columns.intersection(['track_artist','track_name','danceability','energy', 'speechiness', 'acousticness', 'tempo', 'instrumentalness', 'cluster_label'])]
X = df[['danceability', 'energy', 'speechiness', 'acousticness','tempo', 'instrumentalness', 'cluster_label']]
```

```python
X
```

```python
similarity = cosine_similarity(X)
```

```python
similarity
```

```python
np.fill_diagonal(similarity,0)
```

```python
df = pd.DataFrame(similarity, columns=list(en_songs['track_name']))
```

```python
df['id'] = en_songs['track_id']
```

```python
df = df.set_index('id', drop=True)
```
